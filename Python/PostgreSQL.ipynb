```python
import pandas as pd
from sqlalchemy import create_engine

df = pd.read_csv("online_retail_cleaned.csv")
rfm = pd.read_csv("rfm_with_churn.csv")

engine = create_engine("postgresql://postgres:Zxcvbnm,./12@localhost:5432/retail_db")

df.to_sql("retail_transactions", engine, if_exists="replace", index=False, chunksize=500)
rfm.to_sql("customer_rfm", engine, if_exists="replace", index=False, chunksize=500)

print("Done!")
```
