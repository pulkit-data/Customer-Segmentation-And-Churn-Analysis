```python
import pandas as pd
import numpy as np

df = pd.read_csv("online_retail_II.csv", encoding="latin1")
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
print(df.shape)
```

```python
df = df.dropna(subset=["Customer ID"])
print("After dropping null Customer IDs:", df.shape)
```

```python
df = df[~df["Invoice"].astype(str).str.startswith("C")]
print("After removing cancellations:", df.shape)
```

```python
df = df[df["Quantity"] > 0]
df = df[df["Price"] > 0]
print("After removing bad rows:", df.shape)
```

```python
df["TotalPrice"] = df["Quantity"] * df["Price"]
print(df["TotalPrice"].describe())
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
print(df.columns.tolist())
```

```python
print("Final Shape:", df.shape)
print("\nNulls:\n", df.isnull().sum())
df.head()
```

```python
df.to_csv("online_retail_cleaned.csv", index=False)
print("Saved!")
```
